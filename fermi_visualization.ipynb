```python
import os
import random
import matplotlib.gridspec as gs
from astropy.io import fits
import numpy as np
import matplotlib.pyplot as plt
```

```python
pics = []
for i in os.listdir("./data/"):
    if (i[-6:] == "1.fits"):
        pics.append("./data/" + i)
```

```python
for pic in pics:
    
    pic_mm = pic
    pic_dm = pic[:-6] + "2.fits"
    titles = ['model_map', 'deconvolved_map']
    images = [(pic_mm, 2), (pic_dm, 1)]

    i = 0
    fig, axs = plt.subplots(figsize=(14,14),nrows=1, ncols=2, constrained_layout=False)
    for ax in axs.flatten():        
        pic, scale = images[i]
        mm = fits_image = fits.open(pic)[0].data
        mm = mm.astype(np.float32)
        
        ax.imshow(mm, cmap='CMRmap')
        ax.set_title(titles[i], fontsize=28)
        ax.tick_params(labelbottom=False, labelleft=False)

        i = i + 1

    plt.suptitle(pic[pic.find('_')+1:-15], fontsize = 28)        
    plt.show()
    
```
